```python
import gym
import os
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
env = gym.make("CartPole-v0")
```

## Create an anviroments with 5 epiosde and random action

```python
# env.render() is visualiseing the enviroment
# env.reset() is reseting the env and obtain observations
# env.step(action) is applying the action on the enviroment
 
for episode in range(1,6):
    state = env.reset()
    done = False
    score = 0
    
    
    while not done:
        # visualise
        env.render()
        # Defingin the action
        action = env.action_space.sample()
        # applying the action to the env
        n_state, reward, done, info = env.step(action)
        # adding the reward to score
        score += reward
        # printing the number of episode and score
    print("Episode: {} Score: {}".format(episode, score))
env.close()
```

## Model Train

```python
env = gym.make("CartPole-v0")
env = DummyVecEnv([lambda: env])
model = PPO("MlpPolicy",env, verbose=1)
```

```python
# Traing the model 100k times
model.learn(total_timesteps = 100000)
```

## Evaluation

```python
# testing the model 10 times
evaluate_policy(model,env,n_eval_episodes=10,render = True)
# Will show under here the average reward of 200
# We get more info here
```

```python
env.close()
```

## testing the model

```python
for episode in range(1,6):
    obs = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        action,_ = model.predict(obs) # no longer random step instead of random actions
        obs, reward, done, info = env.step(action)
        score+=reward
    print("Episode: {} Score: {}".format(episode, score))
```

```python
env.close()
```
